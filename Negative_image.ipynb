```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img1 = Image.open("Manzara_fotografciligi_2.jpg")
```

```python
from PIL import Image
h,w = img1.size
img = img1.load()
for i in range(0,h):
    for j in range(0,w):
        R,G,B = img[i,j]
        R = 255 - R
        G = 255 - G
        B = 255 - B
        img[i,j] = (R,G,B)
#img1.show()
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img1)
plt.show()
```
